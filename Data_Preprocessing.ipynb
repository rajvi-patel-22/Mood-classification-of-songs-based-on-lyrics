## *REMOVE PUNCTUATION*

```python
#remove punctuations function
def RemovePunctutions(my_str):
    no_punct = ""
    no_punct = re.sub('[!A-Z#?,)(.-:";]', '', my_str)
    return no_punct
```

```python
# Remove puncuation
my_str_series = data_frame['lyrics']
i = 1
for my_str in my_str_series:
    new_str = RemovePunctutions(my_str)
    new_data_frame.set_value(i,'lyrics',new_str)
    i = i+1
  
# print dataframe
new_data_frame.head(5)
```

```python
# do stemming on lyrics and add to dataframe
def generate_stem_words(word):
    suffixes = suffixes = {
    1: [u"ो",u"े",u"ू",u"ु",u"ी",u"ि",u"ा"],
    2: [u"कर",u"ाओ",u"िए",u"ाई",u"ाए",u"ने",u"नी",u"ना",u"ते",u"ीं",u"ती",u"ता",u"ाँ",u"ां",u"ों",u"ें"],
    3: [u"ाकर",u"ाइए",u"ाईं",u"ाया",u"ेगी",u"ेगा",u"ोगी",u"ोगे",u"ाने",u"ाना",u"ाते",u"ाती",u"ाता",u"तीं",u"ाओं",u"ाएं",u"ुओं",u"ुएं",u"ुआं"],
    4: [u"ाएगी",u"ाएगा",u"ाओगी",u"ाओगे",u"एंगी",u"ेंगी",u"एंगे",u"ेंगे",u"ूंगी",u"ूंगा",u"ातीं",u"नाओं",u"नाएं",u"ताओं",u"ताएं",u"ियाँ",u"ियों",u"ियां"],
    5: [u"ाएंगी",u"ाएंगे",u"ाऊंगी",u"ाऊंगा",u"ाइयाँ",u"ाइयों",u"ाइयां"],
}
    for L in 5, 4, 3, 2:
        if len(word) > L + 1:
            for suf in suffixes[L]:
                if word.endswith(suf):
                    return word[:-L]
    return word

newest_data_frame = new_data_frame
for ind in range(1,len(new_data_frame['lyrics'])):
    my_str_series = new_data_frame.loc[ind]['lyrics']
    words_DF = nltk.tokenize.word_tokenize(my_str_series)
    strwithoutstopwords = ""
    stem_word=""
    for str_ in words_DF:
        if str_ not in unique_word_list:
            stem_word=generate_stem_words(str_)
            strwithoutstopwords=strwithoutstopwords+stem_word+" "
    newest_data_frame.set_value(ind,'lyrics',strwithoutstopwords)
    
newest_data_frame
```
